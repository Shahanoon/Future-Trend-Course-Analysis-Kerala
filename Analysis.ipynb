```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

sns.countplot(data=df, x='naac_grade', ax=axes[0], palette='pastel', order=df['naac_grade'].value_counts().index)
axes[0].set_title('NAAC Grade Distribution')
axes[0].tick_params(axis='x', rotation=45)

sentiment_counts = df['sentiment'].value_counts()
axes[1].pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
axes[1].set_title('Google Review Sentiment Analysis')

plt.tight_layout()
plt.savefig('images/analysis_summary.png', bbox_inches='tight')
plt.show()

print("✅ Additional charts saved to 'images/analysis_summary.png'")
```

```python
plt.figure(figsize=(12, 6))
district_counts = df['district'].value_counts()
sns.barplot(x=district_counts.values, y=district_counts.index, palette='magma')
plt.title('Distribution of Colleges by District')
plt.xlabel('Number of Colleges')
plt.ylabel('District')
plt.savefig('images/colleges_by_district.png', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='type_of_college', y='google_rating', palette='coolwarm')
plt.title('Google Rating Distribution by College Type')
plt.xticks(rotation=45)
plt.savefig('images/rating_by_type.png', bbox_inches='tight')
plt.show()
```
